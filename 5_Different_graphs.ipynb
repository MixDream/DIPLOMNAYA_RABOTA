Добавим ещё несколько графиков, которые возможно построить при помощи библиотек **Matplotlib**, **Seaborn** и **Plotly**

**Подготовка данных для графиков и диаграмм**

```python
import pandas as pd

# Ссылка на DataFrame  файла на GitHub
url = 'https://raw.githubusercontent.com/Anna-Lisichka/modul_20_Diploma/main/shopping_data.xlsx'

# Загрузка данных из файла
data = pd.read_excel(url, engine='openpyxl')

# Проверка первых нескольких строк данных
data.head()
```

**1. Библиотека Matplotlib**

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline

# нарисуем сетку
fig, ax = plt.subplots()

# раскрасим поля таблицы
fig.set_facecolor("red")
ax.set_facecolor("blue")

# отберем данные
# fig.set_facecolor("red")
```

```python
plt.figure(figsize=(10, 6), facecolor='red')

plt.plot(
    data["Дата регистрации"],
    data["Цвет"],
    label="Цвет nо дате",  # Добавляем легенду
    color="white",  # Изменяем цвет линии
    linewidth=2,  # Увеличиваем ширину линии
    marker="o",  # Добавляем круглые маркеры к точкам
    markersize=5,  # Увеличиваем размер маркеров
    linestyle="--",  # Изменяем стиль линии на пунктирный
    alpha=0.7  # Уменьшаем прозрачность линии
)

plt.title("Цвет по дате")  # Добавляем заголовок графика
plt.xlabel("Дата регистрации")  # Добавляем название оси X
plt.ylabel("Цвет")  # Добавляем название оси Y

plt.legend()  # Отображаем легенду
plt.grid(True)  # Добавляем сетку

# Установка цвета фона осей
plt.gca().set_facecolor('grey')

plt.show()
```

**2. Библиотека Seaborn**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

ax = sns.lmplot(x='Возраст', y='Сумма', hue='Пол', data = data, fit_reg=True)
#fit_reg=False отключает регрессионую прямую
#hue='Пол' группирует данные по полу

ax.set_xlabels('Возраст')
ax.set_ylabels('Сумма')
```
